```python
class Stack:
    def __init__(self):
        self.items = []

    # Check if it is empty
    def is_empty(self):
        return len(self.items) == 0
    
    # Push a new item
    def push(self, item):
        self.items.append(item)

    # Pop an item
    def pop(self):
        if self.is_empty():
            return None
        return self.items.pop()

    # Peek item
    def peek(self):
        if self.is_empty():
            return None
        return self.items[-1]

    # Return the size
    def size(self):
        return len(self.items)

# create stack    
stack1 = Stack()
# push 3 items
stack1.push(5)
stack1.push(10)
stack1.push(15)

#peek at the item
print("Stack peek:", stack1.peek())

# pop at the item
print ("Stack pop:", stack1.pop())

# check the size
print("Stack size:", stack1.size())
```

```python
class Queue:
    def __init__(self):
        self.items = []

    # Check if the queue is empty
    def is_empty(self):
        return len(self.items) == 0

    # Enqueue (add to the end)
    def enqueue(self, item):
        self.items.append(item)

    # Dequeue (remove from the front)
    def dequeue(self):
        if self.is_empty():
            return None   
        return self.items.pop(0)

    # Return the size of the queue
    def size(self):
        return len(self.items)

    
queue1 = Queue()

queue1.enqueue(20)
queue1.enqueue(40)
queue1.enqueue(60)
             
print("Queue dequeue:", queue1.dequeue()) 
print("Queue size:", queue1.size())        
print("Is empty:", queue1.is_empty())

    
```
