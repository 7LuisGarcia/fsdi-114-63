# Class 2 Stacks and Queues

## Stacks 
A stacks is an ordered collecctions of elements where items are added and removed frm one end, commonly referred to as the top, the ordering principle is  referred to as LIFO (Last In, First Out)

## Queues
A queue is an ordered collecting of elements where items are addad and removed from **opposite** commonly referred to as the "front" and "back" or "rear" of the queue. The ordering principle is referred to as FIFO (First In, First Outor "firts come first served)
