```python

        
# Challenge 1:
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        
# Challenge 2
class LinkedList:
    def __init__(self):
        self.head = None

    # Challenge 3 append method
    def append(self, value):
        new_node = Node(value)

        if self.head is None:
            self.head = new_node
            return

        current = self.head
        while current.next is not None:
            current = current.next

        current.next = new_node

    # Challenge 3.1 print_list method
    def print_list(self):
        current = self.head
        
        if current is None:
            print("List is empty!")
            return

        while current is not None:
            print(current.value)
            current = current.next

    # Count nodes
    def count_nodes(self):
        count = 0
        current = self.head
        while current is not None:
            count += 1
            current = current.next
        return count

    # Insert at beginning
    def prepend(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node

    #challenge 5 removed method
    def remove(self, value):
        # Remove the first node that has the given value
        # If the list is empty nothing to do
        if self.head is None:
            return   

        # if the node to remove is the head
        if self.head.value == value:
            self.head = self.head.next
            return

        current = self.head
        while current.next is not None:
            if current.next.value == value:
                # Skip the node with the matching value
                current.next = current.next.next
                return
            current = current.next

# Create a singly linked list and append 5 values
ll = LinkedList()

ll.append(10)
ll.append(15)
ll.append(20)
ll.append(25)
ll.append(30)
ll.remove(25)
ll.append(40)

print("Initial singly linked list (after append):")
ll.print_list()
print("Node count:", ll.count_nodes())

# Testing prepend method
ll.prepend(50)
print("\nAfter prepend 50:")
ll.print_list()
print("Node count:", ll.count_nodes())
```

```python
# Challenge 1: Node class
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None


# Challenge 2: Doubly Linked List
class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # Challenge 3: append method
    def append(self, value):
        new_node = Node(value)

        # If list is empty
        if self.head is None:
            self.head = new_node
            return

        # Otherwise walk to the end
        current = self.head
        while current.next is not None:
            current = current.next

        # Link last node to new node
        current.next = new_node
        new_node.prev = current

    # Challenge 4: print forward
    def print_forward(self):
        current = self.head

        if current is None:
            print("List is empty!")
            return

        while current is not None:
            print(current.value)
            current = current.next

    # Challenge 5: print backward
    def print_backward(self):
        current = self.head

        if current is None:
            print("List is empty!")
            return

        # Step 1: go to the tail
        while current.next is not None:
            current = current.next

        # Step 2: walk backward
        while current is not None:
            print(current.value)
            current = current.prev


# Create list and test
dll = DoublyLinkedList()

dll.append("A")
dll.append("B")
dll.append("C")
dll.append("D")

print("Forward:")
dll.print_forward()

print("\nBackward:")
dll.print_backward()
```
