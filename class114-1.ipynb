# Pseudocode: Find the Largest Number in a List

1. Start
2. Set `largest` to the first element of the list
3. For each `number` in the list:
   - If `number` > `largest`:
     - Set `largest` = `number`
4. Output `largest`
5. End


```python
nums = [12,4,27,3,1000,50,7]
def find_largest(nums):
    if not nums:
        return "No values to compare"
    largest = nums[0]

    for n in nums[1:]:
        if n > largest:
            largest = n


    return largest
print(find_largest(nums))
```

 class 1 assignment
 Given [3,4,10,15,999,2] Then the ouput should be: total=3
 Create a function that receives a collection and return an integer with the total
Hint
- '%' is the mathemic operand to get the modulus from a division.
- 2 % 2 = 0

```python
def count_evens(numbers):
    total = 0
    for n in numbers:
        if n % 2 == 0:
            total += 1
    return total

numbers = [3, 5, 27, 20, 6, 3.5, 60]
count_evens(numbers)
```
